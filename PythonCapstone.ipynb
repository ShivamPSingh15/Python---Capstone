```python
import pandas as pd
import numpy as np
```

```python

#TASK 1

#Creating Series of lists to use them as columns for the dataframes

id = pd.Series(['A001','A002','A003','A004','A005'])
project = pd.Series(
    ['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5', 'Project 6', 'Project 7', 'Project 8', 'Project 9',
     'Project 10', 'Project 11', 'Project 12', 'Project 13', 'Project 14']
)
name = pd.Series(["John Alter", "Alice Luxumberg", "Tom Sabestine", "Nina Adgra", "Amy Johny"])
city = pd.Series(["Paris", "London", "Berlin", "New York", "Madrid"])
age = pd.Series([25, 27, 29, 31, 30])
status = ["Finished", "Ongoing", "Failed"]
designation_level = [2,2,3,2,3]
status_list = pd.Series(["Finished", "Ongoing", "Finished", "Ongoing", "Finished", "Failed", "Finished", "Failed", 
                         "Ongoing", "Finished", "Failed", "Ongoing", "Finished", "Finished"])
```

```python
# Creating the Employee DataFrame

emp_df = pd.DataFrame({
    "ID" : id, 
    "Name" : name,
    "City" : city,
    "Age" : age,
})
emp_df
```

```python
# Creating the Seniority DataFrame

seni_df = pd.DataFrame({
    "ID" : id, 
    "Designation Level" : designation_level
})

seni_df
```

```python
# Creating the Project DataFrame

project_df = pd.DataFrame({
    "ID" : pd.Series(['A001', 'A002', 'A003', 'A004', 'A005', 'A002', 'A005', 'A003', 'A001', 'A003', 'A001', 'A004', 'A004', 'A005']),
    "Project" : project, 
    "Cost" : pd.Series([1002000, 2000000, 4500000, 5500000, '', 680000, 400000, 350000,'', 300000, 2000000, 1000000, 3000000, 200000]),
    "Status" : pd.Series(["Finished", "Ongoing", "Finished", "Ongoing", "Finished", "Failed", "Finished", "Failed", "Ongoing", "Finished", 
                          "Failed", "Ongoing", "Finished", "Finished"])
})
project_df
```

```python
# Saving the DataFrames into .csv files

emp_df.to_csv("employee_data.csv")
seni_df.to_csv("seniority_data.csv")
project_df.to_csv("project_data.csv")
```

```python
#TASK 2
# Importing the project file as dataframe

project = pd.read_csv("project_data.csv", index_col=0)
project
```

```python
#Creating a list of running averages using 'Cost' column in the project dataframe

running_avg = pd.Series(project['Cost'].rolling(window=3, center=True, min_periods=1).mean())
running_avg = running_avg.astype(float).round()
running_avg
```

```python
#inserting the running average Series as a column in the dataframe beside the 'Cost' column.

project.insert(3,"RunAvgCost", running_avg)
project
```

```python
#Using fillna() backward fill method to fill NaN values with the running averages.

df = project.fillna(method='bfill', axis=1)
df
```

```python
#Since we no longer need the running averages, I chose to drop that column from the dataframe.

df.drop('RunAvgCost',inplace = True, axis=1)
project_new = df
project_new
```

```python
#TASK 3

#Importing 'employees' .CSV file and creating a dataframe using read_csv function.

employees = pd.read_csv("employee_data.csv", dtype = {'Age' : int}, index_col=0)
employees
```

```python
#First, I created a list of the names of the employees
#Next, I used the lamba funcn to split the name into first and second names
#Then, I inserted the two lists into the dataframe
#Finally, dropped the 'Name' column.

names = list(employees['Name'])
names

first_names = list(map(lambda x: x.split()[0], names))
second_names = list(map(lambda x: x.split()[1], names))
employees.insert(1, "First Name", first_names)
employees.insert(2, "Second Name", second_names)
employees.drop('Name', inplace = True, axis=1)

emp_new = employees
emp_new
```

```python
#TASK 4

#Importing 'seniority' .CSV file and creating a dataframe using read_csv function.

seniority = pd.read_csv("seniority_data.csv", index_col=0)
seniority
```

```python
#Merging the Seniority dataframe with the Employee dataframe

pre_final_df = pd.merge(emp_new, seniority, on="ID")
pre_final_df
```

```python
#Creating a final dataframe with the needed data of the employees 


final = pd.merge(pre_final_df, project_new, on='ID')
final
```

```python
#TASK 5
#Using 'for' loop to creating a list of bonus cost
#Later, inserted the list into the final dataframe

results_list=[]

for i, r in final.iterrows():
    if(final['Status'].iat[i] == 'Finished'):
        result=(r['Cost']) + (r['Cost'])*0.05
        results_list.append(result)
    else:
        result=(r['Cost']) + (r['Cost'])*0
        results_list.append(result)

final.insert(8, "Bonus-5%", results_list)
```

```python
#Bonus 5% has been added to the cost of employees who have finished their projects

final
```

```python
#TASK 6

#As per the task given, I used 'for' loop to create a list of new designant ions
#Later, inserted the list into the final dataframe

new_des = []

for i, r in final.iterrows():
    if(final["Status"].iat[i] == 'Failed'):
        result=(r['Designation Level']) - 1
        new_des.append(result)
    else:
        result=(r['Designation Level'])-0
        new_des.append(result)

final.insert(5, 'Designation', new_des)
```

```python
final.drop('Designation Level', inplace = True, axis = 1)
```

```python
#DataFrame with updated designation levels

final.rename(columns={'Designation': 'Designation Level'}, inplace=True)
final
```

```python
#TASK 7

#Using 'for' loop to create a list of genders according to the indices of the 'First Name' column
#Later, inserted the list into the final dataframe



gender = []

for i, r in final.iterrows():
    if(final['First Name'].iat[i] == 'John'):
        result= 'M'
        gender.append(result)
    elif(final['First Name'].iat[i] == 'Tom'):
        result= 'M'
        gender.append(result)
    else:
        result= 'F'
        gender.append(result)
        
```

```python
final.insert(3, 'Gender', gender)
final
```

```python
#Using the same logic as shown above I added prefix to the names.

for i, r in final.iterrows():
    if(final['Gender'].iat[i] == 'M'):
        final['First Name'].iat[i] = 'Mr.' + final['First Name'].iat[i]
    else:
        final['First Name'].iat[i] = 'Mrs.' + final['First Name'].iat[i]

final
```

```python
#TASK 8

#As per the task given, I used 'for' loop to create a list of new designantions
#Later, inserted the list into the final dataframe

new_des = []

for i, r in final.iterrows():
    if(final["Age"].iat[i] > 29):
        result=(r['Designation Level']) + 1
        new_des.append(result)
    else:
        result=(r['Designation Level'])+0
        new_des.append(result)

final.insert(6, 'Designation', new_des)
final.drop('Designation Level', inplace = True, axis = 1)
final.rename(columns={"Designation":"Designation Level"}, inplace = True)
final
```

```python
#TASK 9

#Firstly, I grouped the 'final' dataframe using ID and First Name columns

group = final.groupby(['ID','First Name'])
group
```

```python
#Used sum() function to aggregate the group to find out the total cost of each employee's project

group.sum(['Cost'])
```

```python
#New dataframe: This dataframe has total cost of all projects for each employee

TotalProjCost = group.sum(['Cost'])
TotalProjCost.drop(['Age', 'Designation Level'], inplace = True, axis=1)
TotalProjCost.rename(columns={'Bonus-5%':'Total Cost'}, inplace = True)
TotalProjCost
```
